```python
# Check for duplicated rows
df = pd.read_csv('data.csv')
duplicated_rows = df[df.duplicated()]

# Print the duplicated rows, if any
if len(duplicated_rows) > 0:
    print("Duplicated rows:")
    print(duplicated_rows)
else:
    print("No duplicated rows found.")
```

```python
# remove duplicate rows
df.drop_duplicates
```

```python
# Remove duplicates and count the remaining rows
df = df.drop_duplicates()
num_rows = df.shape[0]

# Print the number of remaining rows
print("Number of rows after removing duplicates: ", num_rows)
```

```python
df.dropna(inplace=True)
df.describe()
```

```python
#Drop irrelevant columns
df.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1, inplace=True) 

print(df.head())

```

```python
# rename
df.rename(columns={"Engine HP": "HP", "Engine Cylinders": "Cylinders", "Transmission Type": "Transmission", "Driven_Wheels": "Drive Mode","highway MPG": "MPG-H", "city mpg": "MPG-C", "MSRP": "Price" }, inplace=True)

# Print the updated DataFrame
print(df.head())
```

```python
#Detect outliers using boxplot for price, Hp and Cylinders sns.boxplot(x=df['Price'])
import seaborn as sns

# Create a boxplot for the 'Price' variable
sns.boxplot(x=df['Price'])
sns.boxplot(x=df['HP'])
sns.boxplot(x=df['Cylinders'])
```

```python
#Plot different features against one another (scatter) 

# Plot  HP against Drive Mode
df.plot.scatter(x='HP', y='Drive Mode')
plt.xlabel('HP')
plt.ylabel('Drive Mode')
plt.title('HP vs. Drive Mode')

# Plot Transmission against MPG-C
df.plot.scatter(x='Transmission', y='MPG-C')
plt.xlabel('Transmission')
plt.ylabel('MPG-C')
plt.title('Transmission vs. MPG-C')

# Show the plots
plt.show()
```

```python
#Plot different features using histogram
# Plot the frequency distribution of HP
df['HP'].plot.hist()
plt.xlabel(' HP')
plt.ylabel('Frequency')
plt.title('HP Frequency Distribution')

# Plot the frequency distribution of Cylinders
df['Cylinders'].plot.hist()
plt.xlabel('Cylinders')
plt.ylabel('Frequency')
plt.title('Cylinders Frequency Distribution')

# Show the plots
plt.show()
```

```python
#Plot different features using heatmap
# Calculate the correlation matrix
corr_matrix = df.corr()

# Create a heatmap of the correlation matrix
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')

# Show the plot
plt.show()
```
