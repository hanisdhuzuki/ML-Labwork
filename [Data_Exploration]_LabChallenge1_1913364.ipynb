```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('AdvWorksCusts.csv')
```

```python
print(df)
```

```python
#replace missing values
df = df.fillna(method='ffill')

#remove duplicate 
df.drop_duplicates(subset = 'CustomerID', inplace = True)
```

```python
#calc summary and desc statistic
summary_stats = df.describe()
print(summary_stats)

desc_stats = df.groupby('PostalCode')['Gender'].describe()
print(desc_stats)
```

```python
#calc correllation
df.corr()

#data visual
sns.scatterplot(x = 'Gender', y = 'YearlyIncome', data = df)
plt.title('Gender vs. YearlyIncome')
plt.show()

sns.scatterplot(x = 'Occupation', y = 'NumberChildrenAtHome', data = df )
plt.title('Occupation vs NumberChildrenAtHome')
plt.show
```

```python
columns_for_heatmap = ['HomeOwnerFlag','NumberCarsOwned', 'NumberChildrenAtHome']
heat_map = sns.heatmap(df[columns_for_heatmap])
plt.show()
```
